# MNIST Classification Demo

This notebook demonstrates how the implemented classifiers work on the MNIST dataset.

Implemented models:
- Random Forest
- Feed-Forward Neural Network
- Convolutional Neural Network

The demo shows:
- loading the dataset
- training models
- making predictions
- evaluation
- edge cases

```python
from data.mnist_loader import load_mnist
from utils.visualization import show_misclassified_examples
from classifier import mnist_classifier
from sklearn.metrics import confusion_matrix, classification_report
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

### Load MNIST dataset

```python
X_train, X_test, y_train, y_test = load_mnist()

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
```

### Example images from dataset

```python
fig, axes = plt.subplots(1,5, figsize=(10,3))

for i in range(5):
    axes[i].imshow(X_train[i].reshape(28,28), cmap="gray")
    axes[i].set_title(y_train[i])
    axes[i].axis("off")

plt.show()
```

<center>

## Random Forest

</center>

#### Train model

```python
rf_model = mnist_classifier.MnistClassifier("rf")

rf_model.train(X_train, y_train)
```

#### Predictions

```python
rf_pred = rf_model.predict(X_test)

print("First 10 predictions:")
for i in range(10):
    print(f"Prediction: {rf_pred[i]}, True label: {y_test[i]}")
```

#### Evaluation

```python
print(classification_report(y_test, rf_pred))
```

#### Confusion matrix

```python
cm = confusion_matrix(y_test, rf_pred)

sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title("Random Forest Confusion Matrix")
plt.show()
```

#### Misclassified examples

```python
show_misclassified_examples(X_test, y_test, rf_pred)
```

# Edge cases

#### Single image prediction
Check that the model can make a prediction for a single image, not only for a full dataset.

```python
pred = cnn_model.predict(X_test[0].reshape(1,-1))

print(f"Prediction: {pred[0]}")
print(f"True label: {y_test[0]}")
```

#### Random noise
Test how the model behaves when the input is random noise instead of a real digit.

```python
noise = np.random.rand(1,784)

pred = cnn_model.predict(noise)

plt.imshow(noise.reshape(28,28), cmap="gray")
plt.title(f"Prediction for noise: {pred[0]}")
plt.axis("off")
```

#### Invalid input
Check how the model handles invalid or empty input.

```python
try:
    cnn_model.predict([])
except Exception as e:
    print(e)
```

#### Prediction shape
Verify that the prediction output has the expected shape.

```python
cnn_model.predict(X_test[:5]).shape
```
